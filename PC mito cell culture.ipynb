```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(5,5),dpi=300) #required to save as .eps

working_directory = "/Users/avellaneda/Desktop/Mitochondria microtubules paper/Figure7/figure/Type3_noc/"
fname = working_directory + "plot_data_normalised.csv"

data = np.loadtxt(fname, delimiter=";", skiprows=1)


x = data[:,1]
i = np.size(data, 1)


t0 = data[:, 2]
t4 = data[:, 3]
t10 = data[:, 4]
t20 = data[:, 5]
t30 = data[:, 6]
t40 = data[:, 7]



for col in np.arange(i):
    if col > 1:
        y = data[:, col]
        #plt.scatter(x, y) # you can comment in if points are needed
        legend = col - 1
        plt.plot(x, y, label = str(legend) + "min")
```

```python
from scipy.signal import lfilter

n = 40             # larger n gives smoother curves
b = [1.0 / n] * n  # numerator coefficients
a = 1# denominator coefficient
y_t0_smooth = lfilter(b, a, t0)
y_t4_smooth = lfilter(b, a, t4)
y_t10_smooth = lfilter(b, a, t10)
y_t20_smooth = lfilter(b, a, t20)
y_t30_smooth = lfilter(b, a, t30)
y_t40_smooth = lfilter(b, a, t40)

plt.plot(x[40:], y_t0_smooth[40:])
plt.plot(x[40:], y_t4_smooth[40:])
plt.plot(x[40:], y_t10_smooth[40:])
plt.plot(x[40:], y_t20_smooth[40:])
plt.plot(x[40:], y_t30_smooth[40:])
plt.plot(x[40:], y_t40_smooth[40:])



plt.xlabel("Distance in microns")
plt.ylabel("Photoconverted Dendra intensity (arbitrary units)")
plt.title('Mitochondria movement')
plt.legend()
plt.ylim([0,2])

plt.savefig(working_directory + "_plot.eps", format = "eps") #if you want to save the plot to the save location of data uncomment line

plt.show() # always needs to be at the end, after plt.savefig() otherwise fig is not saved correctly
```
